```python
with h5py.File('../../Mtot0.95/0.87_0.08_doubledet.h5', 'r') as f:
    print(list(f.keys()))
    Lnu = np.array(f['Lnu'])
    t = np.array(f['time'][:] / 86400)
    nu = np.array(f['nu'][:])
    wv = (c.c / nu / u.Hz).to('angstrom').value
```
