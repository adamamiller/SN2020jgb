- data are in marion_dat format

- each file has 8 columns: l, f, er, ff, lz, fn, ern, ffn

    - l=observed wavelength (angstroms)
    - f=raw flux (ergs/s/cm^2/A)
    - er=error estimate by pixel
    - ff= Fourier transform smoothed spectrum
    - lz=wavelength shifted to rest frame of host (lz=l/(1+z))
    - fn, ern, & ffn are the normalized raw data (f, ff, & er)
      - normalization makes ffn=1.0 at 1.0 microns for NIR data
      - normalization makes ffn=1.7 at 7300 A for optical data

For most analysis it is easiest to look at the normalized, smoothed data (ffn) plotted against the redshift  adjusted wavelength (lz).

- IRTF data in ascii format
- Columns= lambda, de-redshift_lambda, flux, errors
- Units= microns, microns, ergs/s-1/cm-2/A-1, ergs/s-1/cm-2/A-1

```python
# SN metadata

z_20jgb = 0.0307
t_max = 2458986.91
t_max_mjd = t_max - 2400000.5
ebv = 0.404
rv = 3.1

GNIRS = np.loadtxt('./SN2020jgb/ZTF20aayhacx_20200609_Gemini-N_v1.ascii')
snr = np.argwhere(GNIRS[:, 1] > 3 * GNIRS[:, 2]).flatten()
wv_G = GNIRS[snr, 0] / 10000 / (1 + z_20jgb)
flux_G = GNIRS[:, 1] * 10**(0.4 * calALambda(GNIRS[:, 0], rv, ebv))
flux_G = flux_G[snr]

P200 = np.loadtxt('./SN2020jgb/ZTF20aayhacx_20200610_P200_v2.ascii')
P200[:, 2] = P200[:, 2]**.5
P200 = data_binning(P200, size=10)
snr = np.argwhere(P200[:, 1] > 3 * P200[:, 2]).flatten()
wv_P = P200[snr, 0] / 10000 / (1 + z_20jgb)
flux_P = P200[:, 1] * 10**(0.4 * calALambda(P200[:, 0], rv, ebv))
flux_P = flux_P[snr]

arg1 = np.argwhere((wv_G > .85) & (wv_G < .88)).flatten()
arg2 = np.argwhere((wv_P > .85) & (wv_P < .88)).flatten()
flux_P /= (flux_P[arg2].mean() / flux_G[arg1].mean())
```

```python
files = glob.glob('NIR_Marion09/*.flm')
files_other = glob.glob('NIR_Marion09/12fr*.ascii')
files = np.append(files, files_other)
files.sort()
Phase = []
SN = np.array([], dtype=object)
wv, flux, flux_unc = [], [], []
for f in files:
    sp = np.loadtxt(f)
    with open(f, 'r') as sp_f:
        while True:
            l = sp_f.readline()
            if 'Phase=' in l:
                Phase = np.append(Phase, float(l[l.find('=') + 2:-2]))
            if 'E(B-V)=' in l:
                ebv = float(l[l.find('=') + 2:])
            if 'Redshift=' in l:
                z = float(l[l.find('=') + 2:])
            if not '#' in l:
                break
    try:
        arg = sp[:, 2] > sp[:, 3] * 3
        wv.append(sp[arg, 1])
        flux.append(sp[arg, 2])
        flux_unc.append(sp[arg, 3])
    except:
        if sp[0, 0] > 1000:
            wv.append(sp[:, 0] / 1e4 / (1 + z))
        else:
            wv.append(sp[:, 0] / 1e4 / (1 + z))
        E = calALambda(wv[-1] * 10000, 3.1, ebv)
        flux.append(sp[:, 1] * 10**(0.4 * E))
        flux_unc.append(sp[:, 1] * 2e-1 * 10**(0.4 * E))
    index = f.find('on09/')
    f = f[index:]
    SN = np.append(SN, f[5:f.find('_')])
wv = np.array(wv, dtype=object)
flux = np.array(flux, dtype=object)
flux_unc = np.array(flux_unc, dtype=object)
```

```python
norm_G = flux_G[np.abs(wv_G - 1) < 0.05].mean()

depth = 1.6
arg_p = np.argsort(Phase)
arg_l = np.argwhere((Phase[arg_p] > -4) & (Phase[arg_p] < 2)).flatten()
plt.figure(figsize=(9, 1.5 * len(arg_l)))
for k in range(len(arg_l)):
    norm = flux[arg_p][arg_l[k]][np.abs(wv[arg_p][arg_l[k]] - 1) < 0.05].mean()
    x = wv[arg_p][arg_l[k]]
    y = flux[arg_p][arg_l[k]] / norm
    z = flux_unc[arg_p][arg_l[k]] / norm
    X = data_binning(np.array([x, y, z]).T, size=0.0015)
    plt.plot(X[:, 0], X[:, 1] - depth * k, color='k', linewidth=1)
    if k == len(arg_l) - 1:
        plt.plot(wv_G, flux_G / norm_G - depth * (k+0.8), color='crimson', alpha=0.8)
        #plt.plot(wv_P, flux_P / norm_G - depth * k, color='orange', alpha=0.5)
    if Phase[arg_p][arg_l[k]] >= 0:
        p = '+{:.0f}'.format(Phase[arg_p][arg_l[k]])
    else:
        p = '{:.0f}'.format(Phase[arg_p][arg_l[k]])
    plt.text(1.2,
             .8 - depth * k,
             '{}, {}'.format(SN[arg_p][arg_l[k]], p),
             ha='right')
plt.ylim( - depth * (len(arg_l)), 1 + depth * 1.2)
plt.xlim(0.78, 1.21)
plt.yticks([])
plt.xlabel('$\lambda_\mathrm{rest}\ [\mu\mathrm{m}]$')
plt.ylabel(
    '$\mathrm{F_\lambda/F_\lambda(1.0\ \mu\mathrm{m})}+\mathrm{Constant}$')

beta = 0.04
beta2 = 0.097
plt.axvline(0.9227 * ((1 - beta) / (1 + beta))**.5,
            linestyle='--',
            color='0.6')
plt.axvline(1.0927 * ((1 - beta) / (1 + beta))**.5, linestyle='--', color='0.6')
plt.axvline(1.0927 * ((1 - beta2) / (1 + beta2))**.5, linestyle='--', color='crimson')
plt.axvline(0.9227 * ((1 - beta2) / (1 + beta2))**.5, linestyle='--', color='crimson')
plt.text(1.07, depth, 'Mg II')
plt.gca().xaxis.set_minor_locator(MultipleLocator(0.02))
plt.savefig('./../../../../Downloads/NIR_spec_MgII.pdf', bbox_inches='tight')
```
